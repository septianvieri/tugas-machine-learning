1.  Machine learning adalah cabang dari kecerdasan buatan (AI) yang fokus pada pengembangan algoritma dan teknik yang memungkinkan komputer untuk belajar dari dan membuat prediksi atau keputusan berdasarkan data. Sistem machine learning mempelajari pola dan wawasan dari data yang diberikan dan meningkatkan kinerjanya seiring waktu, dan tidak di diprogram secara eksplisit untuk melakukan tugas tertentu.
    Secara singkat, machine learning adalah tentang memberikan kemampuan kepada komputer untuk belajar dari pengalaman (data) tanpa harus diprogram secara eksplisit untuk setiap tugas tertentu.

3. Taxonomi yang ada dalam penerapan machine learning antara lain:  
- Berdasarkan Tipe Pembelajaran:  

        Supervised Learning: Algoritma dilatih menggunakan data berlabel. Tujuannya adalah untuk memprediksi atau mengklasifikasikan data baru berdasarkan pembelajaran dari data berlabel ini.  
        Contoh: Regresi linear, pohon keputusan, random forest, support vector machine (SVM).  
        
        Unsupervised Learning: Algoritma dilatih menggunakan data yang tidak berlabel. Tujuannya adalah untuk menemukan pola atau struktur yang tersembunyi dalam data.  
        Contoh: Klasterisasi (k-means, DBSCAN), pengurangan dimensi (PCA, t-SNE).

        Semi-supervised Learning: Kombinasi antara supervised dan unsupervised learning, menggunakan sejumlah kecil data berlabel dan sejumlah besar data tidak berlabel.  
        Contoh: Algoritma semi-supervised SVM.  

        Reinforcement Learning: Algoritma belajar melalui interaksi dengan lingkungan, menerima umpan balik berupa reward atau penalti, dan bertujuan untuk memaksimalkan reward total.  
        Contoh: Q-learning, deep Q-networks (DQN), Proximal Policy Optimization (PPO).  
        
- Berdasarkan Tipe Data yang Digunakan:

        Continuous Data (Data Kontinu): Algoritma yang digunakan untuk memprediksi variabel kontinu.  
        Contoh: Regresi linear, regresi ridge.

        Categorical Data (Data Kategorikal): Algoritma yang digunakan untuk mengklasifikasikan data ke dalam kategori yang berbeda.
        Contoh: Klasifikasi pohon keputusan, naive Bayes, logistic regression.  

- Berdasarkan Tujuan atau Tugas:

        Classification (Klasifikasi): Memetakan input ke dalam salah satu dari beberapa kategori.  
        Contoh: Klasifikasi email sebagai spam atau bukan spam, pengenalan wajah.
        
        Regression (Regresi): Memprediksi nilai kontinu.  
        Contoh: Prediksi harga rumah, estimasi pendapatan.

        Clustering (Klasterisasi): Mengelompokkan data ke dalam kelompok atau klaster berdasarkan kesamaan.  
        Contoh: Segmentasi pelanggan, analisis pasar.

        Dimensionality Reduction (Pengurangan Dimensi): Mengurangi jumlah fitur dalam dataset sambil mempertahankan informasi yang paling penting.  
        Contoh: Principal Component Analysis (PCA), Linear Discriminant Analysis (LDA).  

- Berdasarkan Metode Pembelajaran:

        Instance-based Learning: Algoritma menyimpan contoh data dan memprediksi berdasarkan contoh serupa di masa lalu.  
        Contoh: K-nearest neighbors (KNN).

        Model-based Learning: Algoritma membangun model yang dapat membuat prediksi berdasarkan data input.  
        Contoh: Regresi linear, neural networks.  

- Berdasarkan Teknik yang Digunakan:

        Linear Models: Model yang mengasumsikan hubungan linear antara input dan output.  
        Contoh: Regresi linear, regresi logistik.

        Tree-based Models: Model yang menggunakan struktur pohon untuk membuat keputusan atau prediksi.  
        Contoh: Pohon keputusan, random forest, gradient boosting trees.

        Neural Networks and Deep Learning: Model yang menggunakan jaringan saraf tiruan untuk menangani masalah kompleks dengan banyak lapisan pemrosesan.  
        Contoh: Multi-layer perceptrons (MLP), convolutional neural networks (CNN), recurrent neural networks (RNN).

1) Sempurnakan Data Set Universitas Esa Unggul memiliki detail mahasiswa sebagai berikut :  
Ani, Budi, Joni, Jono dan Lono Setiap mahasiswa dating ke kampus mulai dari hari senin sampai
minggu Untuk melakukan kegiatan kuliah Ani menghabiskan biaya 30000 dikampus setiap datang
Budi menghabiskan biaya 35000 dikampus setiap datang Joni menghabiskan biaya 20000
dikampus setiap datang Jono 25000 dan Lono 15000

    

    a) Berapa rata-rata mahasiswa datang pada minggu ini?  
    b) Kapan biaya tertinggi terjadi?  
    c) Hari apa biaya lebih dari 110000?  
    d) Siapa yang paling banyak datang ke kampus?  
    e) Siapa yang datang pada hari minggu?  
    f) Berapa biaya tertinggi dan terendah?  
    g) Berapa frekuensi datang tertinggi dan terendah?  

````python
import pandas as pd

data = {
    'Hari': ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'],
    'Datang': [2, 3, 4, 1, 1, 5, 2],
    'Biaya': [30000*2, 35000*3, 25000*4, 15000*1, 20000*1, 30000*5, 35000*2],
    'Mahasiswa': ['Ani', 'Budi', 'Jono', 'Lono', 'Joni', 'Ani', 'Budi']
}

dataset = pd.DataFrame(data)
rata_rata_datang = dataset['Datang'].mean()
biaya_tertinggi_hari = dataset.loc[dataset['Biaya'].idxmax(), 'Hari']
biaya_tertinggi = dataset['Biaya'].max()
hari_biaya_lebih_110000 = dataset[dataset['Biaya'] > 110000]['Hari'].tolist()
mahasiswa_terbanyak = dataset['Mahasiswa'].value_counts().idxmax()
datang_minggu = dataset.loc[dataset['Hari'] == 'Minggu', 'Mahasiswa'].values[0]
biaya_terendah = dataset['Biaya'].min()
frekuensi_tertinggi = dataset['Datang'].max()
frekuensi_terendah = dataset['Datang'].min()

print(f"a) {rata_rata_datang:.2f}")
print(f"b) {biaya_tertinggi_hari}")
print(f"c) {', '.join(hari_biaya_lebih_110000)}")
print(f"d) {mahasiswa_terbanyak}")
print(f"e) {datang_minggu}")
print(f"f) Tertinggi: {biaya_tertinggi}, Terendah: {biaya_terendah}")
print(f"g) {frekuensi_terendah}")
````
